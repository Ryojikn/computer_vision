# What is Motion Blur?

Motion blur happens when there's a high amount of light being dragged accross the sensor.
This "drag" makes an effect of a light that seems to be moving, and due to that, it creates a trail on the image.

You can see this effect happening when you "shake" accidentally by taking a picture, for example.
