### Faça a seleção de uma imagem qualquer que você possua no computador

```python
import cv2

im = cv2.imread('raw_imgs/booktable.jpg')
im2 = cv2.imread('raw_imgs/computerdesk.jpg')
```

### Realizando a detecção nas respectivas imagens

```python
import cvlib as cvl

bbox, label, conf = cvl.detect_common_objects(im)
bbox2, label2, conf2 = cvl.detect_common_objects(im2)
```

### Desenhando o que foi detectado de volta nas imagens

```python
from cvlib.object_detection import draw_bbox

output_im = draw_bbox(im, bbox, label, conf)
output_im2 = draw_bbox(im2, bbox2, label2, conf2)
```

### Exibindo o resultado

```python
from PIL import Image

Image.fromarray(output_im)
```

```python
Image.fromarray(output_im2)
```
