# O que são imagens

Antes de mais nada, de falarmos sobre quaisquer detalhes sobre imagens, é importante entendermos o que é uma imagem para o nosso cérebro.

O ser humano, salvo exceções, é tão acostumado a enxergar que raramente pensa sobre como funciona sua própria visão, e ainda mais, o quão otimizada ela pode ser. Quando nos aproximamos do contexto de tecnologia, com tantas câmeras, celulares, novos televisores e videogames, ficamos um pouco mais curiosos sobre o que é uma imagem e como as empresas conseguem lançar tantos aparelhos novos, de qualidade gráfica tão bonita, e "tão realistas", como assim dizemos.

Tudo isso, é dado por um conjunto entre o nosso aparelho visual e o espectro de luzes que conseguimos enxergar.

## O que é o espectro de luz?

Geralmente, quando nos referimos à palavra "luz", estamos direcionando este entendimento apenas ao espectro visível ao olho humano, porém é importante nos recordarmos de que como em ciência de dados, tudo se refere ao espaço amostral em que você está trabalhando.
A luz, nada mais é do que um tipo de onda eletromagnética, e como toda onda, ela possui um comprimento e também uma frequência.

<img src="https://people.whitman.edu/~dunnivfm/FAASICPMS_Ebook/CH1/Figures/F1_2_Electromagnetic_Spectrum.gif">

Antes de continuar o papo sobre imagens, vamos entender o funcionamento de nosso sensor ocular.

## Como funciona a visão humana?

Para que nós possamos enxergar, a luz visível é encaminhada ao ponto em que estamos olhando, e então essa onda é refletida para o nosso olho.

O olho possui a cornea, o cristalino, a iris, e a retina. A retina é onde são formadas as imagens, e ela possui um monte de fotoreceptores, que são classificados em dois tipos, os bastonetes (rods) e os cones.

<img src="https://www.anatomiadocorpo.com/wp-content/uploads/2016/04/olho-humano-globo-ocular.jpg">

Quando a luz chega ao cristalino, o olho busca direcionar a maior parte do tempo toda essa luz para a fóvea, que é onde se concentram a maior quantidade de cones no olho humano.

A fóvea, é onde nós conseguimos enxergar com melhor qualidade, e os cones são muito bons em enxergar em detalhe e em condições de iluminação boas, como na luz do dia. No restante da retina, existe uma concentração menor, quase inexistente de cones, e uma concentração muito grande de bastonetes que também são excelentes sensores, não pra enxergar detalhes, mas na visualização do todo, e também são excelentes em enxergar com condições de iluminação mais baixas.

<img src="https://www.gta.ufrj.br/grad/10_1/retina/figuras/olho.jpg">

Em geral, nós conseguimos enxergar um grande espectro da luz classificada como visível, e conseguimos também enxergar em diversas condições, mas devido a essa diferença entre os bastonetes e os cones, nós não conseguimos enxergar em duas condições extremas ao mesmo tempo.

<img src="https://slideplayer.com/slide/16302671/95/images/18/Grey+Levels+humans+can+see.jpg">

Na imagem acima, podemos ver a camada "Scotopic", em baixas condições de luz, onde geralmente os bastonetes nos auxiliam mais. e na parte superior da camada "Photopic", os cones nos auxiliam mais para enxergar em condições ideais de luz, embora os sensores sejam acionados em conjunto o tempo todo, por diferentes raios de luz que o cristalino recebe.

Entre essas duas partes, temos uma boa camada de adaptação que é o cérebro auxiliando a retina, a como processar as informações recebidas por ambos os sensores para nos permitir enxergar. É por este tempo de adaptação e a diferença entre os dois tipos de fotoreceptores, que ao sair de um ambiente muito iluminado, para uma sala de cinema, por exemplo, o nosso olho leva alguns segundos pra se adaptar. O mesmo acontece se você está num lugar muito escuro e sai para o sol de uma vez.

## Funcionamento de uma câmera

Compreendendo agora como o olho humano é capaz de enxergar, temos uma noção de que para simular a percepção visual do mundo temos sensores que são capazes de captar feixes de luz, e transformar isso em uma determinada cor para o cérebro.

De uma forma geral, as câmeras buscam similar a mesma estrutura que existe no olho, possui uma lente (cristalino), e também o diafragma (iris) que controla o nível de luz que entra no sensor (retina). 

<img src="https://lightartacademy.com/blog/wp-content/uploads/2014/11/How-does-a-camera-work-3.jpg">

Após isso, o microprocessador é responsável por compreender os dados que estão presentes no sensor e compor uma imagem.
Este sensor, é composto de uma matriz com milhares de fotoreceptores que buscam simular os mesmos fotoreceptores que existem na retina, e ele converte essa luz para uma matriz de pontos com valores, onde cada ponto representa um **pixel** (picture element).

<img src="https://media.geeksforgeeks.org/wp-content/uploads/Pixel.jpg">

## Composição de uma imagem

Para que uma imagem seja formada, duas técnicas são aplicadas:

### Amostragem

Lembre que ao criarmos um sensor semelhante a retina, nós não temos exatamente os mesmos sensores que existem na retina humana, estamos criando aqui uma representação deste funcionamento, e portanto, é possível que adaptações sejam necessárias.

Uma destas adaptações é a amostragem (*sampling*), feita quando a luz atinge o sensor, e pegamos apenas uma amostra do que o mundo real nos fornece. Esta amostra é feita no domínio espacial, e busca representar os dados que a matriz de fotoreceptores conseguiu obter.

Perceba que como falamos de amostragem, quanto maior for a matriz de fotoreceptores, maior será a definição da imagem, pois em uma mesma área, consegue-se ter uma riqueza maior de detalhes devido ao fato de cada microreceptor ser capaz de obter um valor ativado por um feixe de luz.

Uma forma legal de mostrar como a amostragem é importante, é quando pensamos no próprio Google Street View e buscamos um local qualquer: uma foto foi tirada pelo Satelite, que possui um sensor designado para uma riqueza de detalhes, e conforme nós vamos dando zoom, essa riqueza de detalhes vai diminuindo, pois a amostragem se torna mais e mais insuficiente.

Outra forma de visualizar é pensar em como os videogames evoluiram, e em um mesmo espaço físico, representa-se uma quantidade maior de cores.

### Quantização

Quando os sensores são ativados, geram um impulso que é medido em milivolts ou ainda mais sensível que isto, e este valor, para que possa ser computado, o processo nomeado de quantização é necessário. Quantização trata-se da maneira de alterar um valor contínuo, cuja amplitude varia entre infinitos valores, para um universo de valores mensurável.

Quando falamos sobre o armazenamento desta informação em 8 bits, então temos um universo que vai de 0 a 255, por exemplo. É importante ressaltar que existem diversas técnicas de quantização, mas não detalharemos sobre isso agora.

Uma das mais comuns faz o seguinte tipo de alteração:

Valor inicial = 17,5

Quantizado = [17/2] (round below) * 2
Quantizado = 16

Perceba que ao colocar em modulo e arredondar para baixo, a informação é levemente distorcida, com isso pode-se economizar a memória ao trabalhar com inteiro ao invés de um valor float em detrimento da leve perda de informação. O processo de quantização é um dos processos usados na compressão de imagens, sem que a perda de informação seja muito grande.

### E os canais de cores?

Se mapearmos apenas de 0 a 255, estaremos falando apenas da representação de cores que vai numa escala de branco até preto, também conhecida como escala de cinza. Nos computadores, usamos o sistema RGB (Red, Green and Blue), o que aumenta a nossa riqueza de detalhes na imagem.

### Tamanho de uma imagem

Agora, perceba, caso tenhamos uma matriz de sensores de 1000x1000, teremos um total de:

    - 24000000 bits = 1000 x 1000 x 3 (canais de cores) x 8 (quantidade de bits)

    - 24000000 bits / 1 byte = 3 000 000 bytes

    - 3000000 bytes / 1024 / 1024 = 2,8610 Mbytes

Ou seja, numa única imagem, já temos 2,86 Mbytes, certo?
E se considerassemos que um segundo de vídeo possui 30 imagens?

    - 30 frames = 85,83 Mbytes

    - 1 minuto de video = 5 149 Mbytes

    - 1 hora de video = 310 Gbytes
