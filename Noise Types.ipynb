# Noise Types

```python
from skimage.util import random_noise
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
dogimg = cv2.imread('imgs/dog.jpg', 0)
plt.imshow(dogimg, cmap='gray')
```

```python
histogram = plt.hist(dogimg.ravel(),256,[0,256])
```

### Gaussian Noise

```python
noisy_img = dogimg + np.random.normal(scale=20, size=dogimg.shape)
```

```python
plt.imshow(noisy_img, cmap='gray')
```

```python
plt.figure(figsize=[10, 3])
plt.subplot(121)
plt.hist(dogimg.ravel(),256,[0,256])
plt.subplot(122)
plt.hist(noisy_img.ravel(),256,[0,256]) 
plt.show()
```

### Rayleigh Noise

```python
## adding rayleigh noise
noisy_img = dogimg + np.random.rayleigh(scale=20, size=dogimg.shape)
```

```python
plt.imshow(noisy_img, cmap='gray')
```

```python
plt.figure(figsize=[10, 3])
plt.subplot(121)
plt.hist(dogimg.ravel(),256,[0,256])
plt.subplot(122)
plt.hist(noisy_img.ravel(),256,[0,256]) 
plt.show()
```

### Gamma Noise

```python
## adding gamma noise
noisy_img = dogimg + np.random.gamma(shape=1, scale=20, size=dogimg.shape)
```

```python
plt.imshow(noisy_img, cmap='gray')
```

```python
plt.figure(figsize=[10, 3])
plt.subplot(121)
plt.hist(dogimg.ravel(),256,[0,256])
plt.subplot(122)
plt.hist(noisy_img.ravel(),256,[0,256]) 
plt.show()
```

### Exponential Noise

```python
## adding gamma noise
noisy_img = dogimg + np.random.exponential(scale=20, size=dogimg.shape)
```

```python
plt.imshow(noisy_img, cmap='gray')
```

```python
plt.figure(figsize=[10, 3])
plt.subplot(121)
plt.hist(dogimg.ravel(),256,[0,256])
plt.subplot(122)
plt.hist(noisy_img.ravel(),256,[0,256]) 
plt.show()
```

### Uniform Noise

```python
## adding gamma noise
noisy_img = dogimg + np.random.uniform(high=20, size=dogimg.shape)
```

```python
plt.imshow(noisy_img, cmap='gray')
```

```python
plt.figure(figsize=[10, 3])
plt.subplot(121)
plt.hist(dogimg.ravel(),256,[0,256])
plt.subplot(122)
plt.hist(noisy_img.ravel(),256,[0,256]) 
plt.show()
```

### Impulse Noise

```python
## adding gamma noise
noisy_img = random_noise(dogimg, mode='s&p')
```

```python
plt.imshow(noisy_img, cmap='gray')
```

```python
plt.figure(figsize=[10, 3])
plt.subplot(121)
plt.hist(dogimg.ravel(),256,[0,256])
plt.subplot(122)
plt.hist(noisy_img.ravel(),256,[0,256]) 
plt.show()
```
